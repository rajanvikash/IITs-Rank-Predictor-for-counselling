# Best Model Selection By Hyper Parameter Tuning...

```python

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score


cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso , LinearRegression
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y_close)
```

# Our Best Model is DecisionTreeRegressor indcated from GridSearch data

```python
model = DecisionTreeRegressor(criterion='mse' , splitter='random')
model.fit(x_train , y_train )
model.score(x_test, y_test)
```

```python
def predict_closerank(year,pool , program_duration, degree ,catergory , institute_name , department_name  ):    
    loc_index_pool = np.where(X.columns==pool)[0][0] # its finds the actual column number of pool in training dataframe
    loc_index_program_duration = np.where(X.columns==program_duration)[0][0]
    loc_index_degree = np.where(X.columns==degree)[0][0]
    loc_index_category = np.where(X.columns==catergory)[0][0]
    loc_index_iit_name = np.where(X.columns==institute_name)[0][0]
    loc_index_department_name = np.where(X.columns==department_name)[0][0]
    

    x = np.zeros(len(X.columns))# initializing  all columns as zero initially 
    x[0] = year  # giving first column as year value as our initial training data set contains first columns as year
    
    
    if loc_index_pool >= 0: # checking either return column number for pool input is valid or not
        x[loc_index_pool] = 1 # if we have valid column number , then set the value of that column as 1 i.e we want to search for
                                # that particular pool , 1 means that out of different pools , select that particular pool
        
    if loc_index_program_duration >= 0:
        x[loc_index_program_duration] = 1
        
    if loc_index_degree >= 0:
        x[loc_index_degree] = 1
    
    if loc_index_category >= 0:
        x[loc_index_category] = 1
        
    if loc_index_iit_name >= 0:
        x[loc_index_iit_name] = 1
        
    if loc_index_department_name >= 0:
        x[loc_index_department_name] = 1
        
        
        
    

    return model.predict([x])[0]    # input is here 2d array
```
